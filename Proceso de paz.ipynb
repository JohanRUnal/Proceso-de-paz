## DATA CLEANING
 Luego de ver como estan estructuradas las variables vamos a hacer un preprocesamiento de los datos que nos permita tener datos mas apropiados antes de implementar un modelo.

```python
# inciamos el proceso modificando los valores de la variable Sexo
# cambiando tods los valores a mayusculas
datos['Sexo'] = datos['Sexo'].replace({'Masculino':'MASCULINO', 'Femenino':'FEMENINO'})
print(datos['Sexo'].value_counts())
```

```python
# para la variable Departamento de residencia vamos 
datos = datos[datos['Departamento de residencia'] != '<No Registra>']
```

```python
''' en la variable referente a los rangos de edades , descartamos aquellos poco significativos como los son los implican edades superiores a los 60 años  '''
datos = datos[datos['Grupo Etario'].isin(['Entre 26 y 40 años', 'Entre 41 y 60 años','Entre 18 y 25 años'])]
datos.head(10)
```

```python
# por otro lado vamos a concentrarnos en los grupos armados FARC, AUC y ELN ya que el número de combatientes de otros grupos es numericamente menos significativo.
datos = datos[datos['Ex Grupo'].isin(['AUC', 'FARC', 'ELN'])]

''' adicionalmente tomaremos solamente el departamento de residencia ya que considerar el municipio puede traer grandes dificultades a los modelos debido a la gran cantidad de clases
lo cual nos puede  llevar a situaciones de underfitting'''
datos = datos.iloc[:,0:6] 
datos.head()
```

### codificando variables

Desafortunadamente *sklearn Decision Tree* no puede manipular variables categoricas, por lo tanto debemos codificar las variables de manera numérica como variables indicadoras.


```python
from sklearn import preprocessing
# codificando la variable "situación frente al proceso"
le_sitproce = preprocessing.LabelEncoder()
le_sitproce.fit(['Culminado', 'Fuera del Proceso','En Proceso', 'Ausente del proceso'])
datos['Situación frente al proceso'] = le_sitproce.transform(datos['Situación frente al proceso'])

# codificando la variable "Tipo de Desmovilizacion"
le_tipdesm = preprocessing.LabelEncoder()
le_tipdesm.fit(['Colectiva', 'Individual'])
datos['Tipo de Desmovilización'] = le_tipdesm.transform(datos['Tipo de Desmovilización'])

# Codificando la variable 'Ex grupo'
le_Exgrupo = preprocessing.LabelEncoder()
le_Exgrupo.fit(['AUC', 'FARC', 'ELN'])
datos['Ex Grupo'] = le_Exgrupo.transform(datos['Ex Grupo'])

# codificando la variable "Sexo"
le_sexo = preprocessing.LabelEncoder()
le_sexo.fit(['MASCULINO', 'FEMENINO'])
datos['Sexo'] = le_sexo.transform(datos['Sexo'])
# Codificando la variable "Grupo Etario"
le_ranedad= preprocessing.LabelEncoder()
le_ranedad.fit(['Entre 26 y 40 años', 'Entre 41 y 60 años', 'Entre 18 y 25 años','Mayor de 60 años'])
datos['Grupo Etario'] = le_ranedad.transform(datos['Grupo Etario'])

# Codificando al variable "Departamento de residencia"
departamentos = datos['Departamento de residencia'].unique()
le_depart = preprocessing.LabelEncoder()
le_depart.fit(departamentos)
datos['Departamento de residencia'] = le_depart.transform(datos['Departamento de residencia'])





```

## MODELO 

Para el conjunto de datos que se tiene se pretende construir un modelo supervisado que nos permita  "perfilar"  a un combatiente y poder predecir basado en la información disponible cual sera su "situación frente al proceso". 

Para ello y dada la naturaleza de las variables se consideran dos modelos: arbol de decisión y el uso de un clasificador bayesiano.

   ### Modelo 1- Arbol de decisión

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Co el find e construir el modelo vamos a separar nuestras variables predictoras de la variable de respuesta.
X = datos.iloc[:,1:] 
y = datos.iloc[:,0] # Variable a predecir "Estado frente al proceso"
```

```python
# para implementar nuestro modelo vamos a fraccionar nuestro data set de modo que tengamos un conjunto de entrenamiento y otro para hacer el test del modelo obtenido
from sklearn.model_selection import train_test_split

# A continuación generaremos los conjuntos de datos X_train, X_test, y_train, y_test.  Tomando como parametro para el test_size = 0.3
# Luego tomaremos el 30 % de los datos para hacer el test de nuestro modelo.
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 3)

# veamos las dimensiones de : X_train,X_test, y_train, y_test.
print('la dimensión de X_train es \n:', X_train.shape)
print('la dimensión de X_test es \n:', X_test.shape)
print('la dimensión de y_train es \n:', y_train.shape)
print('la dimensión de t_test es \n:', y_test.shape)

```

```python
# Para medir la impureza de los nodos obtenidos puede usar como ciretior : indice de Gini, entropy. Usaremos esta ultima para poder medir la gananacia de iinformación en cada nodo(Gain_info)
modelo_arbol = DecisionTreeClassifier(criterion = 'entropy', max_depth=4 )
modelo_arbol # veamos los aprametros por defecto
```

```python
# ahora entrenamos el modelo con los datos de X_train y y_train.
modelo_arbol.fit(X_train,y_train)
```

```python
# veamos como se comporta el modelo con el conjunto de test
arbol_predic = modelo_arbol.predict(X_test)
```

```python
# veamos algunos resultados comparando el comportamiento de nuestro modelo con la variable de respuesta
print(arbol_predic[0:5])
print(y_test[0:5])
```

## Evaluación del modelo

Ahora usando las metricas disponibles en sklearn vamos a medir la precisión de nuestro modelo

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print('El accuracy de nuestro modelo es', metrics.accuracy_score(y_test, arbol_predic))
```
